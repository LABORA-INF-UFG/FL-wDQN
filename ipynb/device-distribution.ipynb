```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def hist(distances):
    plt.hist(distances, bins=20, density=False, alpha=0.7, color='#1e7cd4', edgecolor='white', linewidth=1.5)
    plt.xlabel('Distância', fontsize=22)
    plt.ylabel('Frequência Absoluta', fontsize=22)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tick_params(axis='both', labelsize=20)
    plt.show()
```

```python
def scatter(distances, angles, vetor_tam_dados, label):
    x = distances * np.cos(angles)
    y = distances * np.sin(angles)

    central_point = np.array([0, 0])
    x = np.append(x, central_point[0])
    y = np.append(y, central_point[1])
   
    plt.scatter(x, y, s=vetor_tam_dados, alpha=0.5, color='#05539c')
    plt.scatter(central_point[0], central_point[1], s=400, color='black', edgecolor='black', marker='^', label='Centro')

    plt.xlabel(label, fontsize=22)
    plt.tick_params(axis='both', labelsize=20)
    plt.yticks(np.arange(-500, 501, 250))

    plt.show()
```

```python
np.random.seed(1)
num_points = 100

a, b = 100, 500
distances, angles = a + (b - a) * np.random.rand(num_points), 2 * np.pi * np.random.rand(num_points)

hist(distances)
```

```python
indice = 0

point1 = np.array([0, 0])
x = distances * np.cos(angles)
y = distances * np.sin(angles)
point2 = np.array([x[indice], y[indice]])

d = np.linalg.norm(point2 - point1)
print(d)
print(distances[indice])
```

```python
lista = []
for i in range(1,101):
    df = pd.read_pickle(f"/home/renan/Documentos/FL/datasets/mnist/non-iid-0.9-100-rotation-45/{i}_train.pickle")
    lista.append(len(df))

min, max = 5, 100
tam_mnist = ((lista - np.min(lista)) / (np.max(lista) - np.min(lista)) * (max - min) + min).tolist()
tam_mnist.append(1)
scatter(distances, angles, tam_mnist, 'NIID R-MNIST')
```
