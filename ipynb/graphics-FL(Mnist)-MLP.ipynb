```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as ticker
```

```python
def graph_values_acc(uploads_mean_acm, acc_mean):
    lista_base = list(range(10, 5001, 10))
    indices_maior_ou_igual = [next((i for i, valor in enumerate(uploads_mean_acm) if valor >= x), None) for x in lista_base]
    nova_lista = [valor for valor in indices_maior_ou_igual if valor is not None]
    y = np.array(acc_mean)[nova_lista]
    x = lista_base[:len(y)]
    return x, y
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

nivel_confianca = 0.95
z = stats.norm.ppf(1 - (1 - nivel_confianca) / 2) 
uploads_1_ci = z * (np.std(uploads_1_list, axis=0, ddof=1) / np.sqrt(len(uploads_1_list)))
uploads_2_ci = z * (np.std(uploads_2_list, axis=0, ddof=1) / np.sqrt(len(uploads_2_list)))
uploads_3_ci = z * (np.std(uploads_3_list, axis=0, ddof=1) / np.sqrt(len(uploads_3_list)))
uploads_4_ci = z * (np.std(uploads_4_list, axis=0, ddof=1) / np.sqrt(len(uploads_4_list)))
uploads_5_ci = z * (np.std(uploads_5_list, axis=0, ddof=1) / np.sqrt(len(uploads_5_list)))
uploads_6_ci = z * (np.std(uploads_6_list, axis=0, ddof=1) / np.sqrt(len(uploads_6_list)))

valores = [np.mean(uploads_1_list), np.mean(uploads_3_list), np.mean(uploads_5_list), np.mean(uploads_2_list), np.mean(uploads_4_list), np.mean(uploads_6_list)]
ci_sums = [uploads_1_ci, uploads_3_ci, uploads_5_ci, uploads_2_ci, uploads_4_ci, uploads_6_ci]

fig, ax = plt.subplots(figsize=(3, 3))
categorias = ['FedAvg', 'FedAvg-w$_{MILP}$', 'FedAvg-w$_{DQN}$', 'POC', 'POC-w$_{MILP}$', 'POC-w$_{DQN}$']
largura_barra = 0.6
cores = ['#00aaff', 'orange', 'red', 'green', 'magenta', 'blue']

barras = ax.bar(categorias, valores, yerr=ci_sums, capsize=5, color=cores, edgecolor='white', alpha=0.6, linewidth=3, width=largura_barra)

plt.ylabel('Transmissões \n com Êxito', fontsize=14)
plt.rcParams['legend.fontsize'] = 14
legenda = plt.legend(barras, categorias, loc='lower center', framealpha=0.4, ncol=1)
plt.xticks([0, 1, 2, 3, 4, 5], ['', '', '', '', '', ''], fontsize=14)
plt.show()
```
